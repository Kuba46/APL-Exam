# Вопрос №5. Jupiter Notebook, модули визуализации данных, основные способы визуализации данных.
**Jupyter Notebook** — это интерактивная среда, которая позволяет создавать и совместно использовать документы, содержащие живой код, уравнения, визуализации и повествовательный текст. Она широко используется для анализа данных, научных исследований и обучения. В Jupyter Notebook можно использовать различные модули для визуализации данных.

1. **Matplotlib**:
Основной модуль для создания статических, анимационных и интерактивных визуализаций в Python.
Поддерживает широкий спектр графиков, включая линейные, столбчатые, гистограммы и т.д.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sine Wave')
plt.show()  
```

2. **Seaborn**: Основывается на Matplotlib и предоставляет высокоуровневый интерфейс для создания привлекательных и информативных статистических графиков.
3. **Pandas**: Хотя Pandas в основном используется для манипуляции и анализа данных, он также предоставляет базовые возможности для визуализации данных.

Далее следует пример использования всех трёх выше перечисленных библиотек на примере 2 лабы. 

```python
import seaborn as sns
import matplotlib.pyplot as mplt
import pandas

import warnings
warnings.filterwarnings("ignore")
```

```python
df1 = pandas.read_csv("ships.csv")
df1.head(11)
```

```python
fig = mplt.figure(figsize=(16, 9))
fig.suptitle("Зависимость средней максимальной загрузки кораблей от длины корабля")

ax = sns.lineplot(data=df1, x="length", y="dwt", marker="o", errorbar=None, hue="ship_name")

ax.set_xlabel("Длина корабля (метры)")
ax.set_ylabel("Максимальная нагрузка (тонны)")

ax.legend(loc=2, title="Условные обозначения")

mplt.grid()
mplt.show()
```
